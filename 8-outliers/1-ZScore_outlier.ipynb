## **Approach 2: Calculating the Zscore**
**(xi - u) / sd**  
                                    (sd -> standard deviation)

```python
df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean())/df['cgpa'].std()
```

```python
df.head()
```

## **the data without outliers lies in [-3 to 3]**

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore'] < -3]
```

```python
df[(df['cgpa_zscore'] > 3) | (df['cgpa_zscore'] < -3)]
```

# **Trimming**

```python
new_df = df[(df['cgpa_zscore'] < 3) & (df['cgpa_zscore'] > -3)]
```

```python
new_df
```

# **Capping**

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
lower_limit
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
